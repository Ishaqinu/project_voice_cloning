```python
# Read translated text from file_name.txt
file_path = '/content/speech2.txt'
try:
    with open(file_path,'r') as file:
        file_contents = file.read()
    text_variable = file_contents
except FileNotFoundError:
    print(f"The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred:{str(e)}")

# Pick a "preset mode" to determine quality. Options: {"ultra_fast", "fast" (default), "standard", "high_quality"}. See docs in api.py
preset = "high_quality"

# upload audio clips. They must be a WAV file
CUSTOM_VOICE_NAME = "custom"

custom_voice_folder = f"tortoise/voices/{CUSTOM_VOICE_NAME}"
os.makedirs(custom_voice_folder)
for i, file_data in enumerate(files.upload().values()):
  with open(os.path.join(custom_voice_folder,f'{i}.wav'),'wb') as f:
    f.write(file_data)
```

```python
# Generate speech
voice_samples, conditioning_latents = load_voice(CUSTOM_VOICE_NAME)
max_tokens_per_pass = 400  # Define the maximum number of tokens per inference pass

# Split the text into segments based on the maximum tokens allowed
segments = [text_variable[i:i+max_tokens_per_pass] for i in range(0,len(text_variable),max_tokens_per_pass)]

generated_segments = []

for segment in segments:
    gen = tts.tts_with_preset(segment, voice_samples=voice_samples, conditioning_latents=conditioning_latents,preset=preset)
    generated_segments.append(gen)

# Concatenate the generated segments into a single audio
generated_audio = torch.cat(generated_segments,dim=-1)

# Save and play the generated audio
torchaudio.save(f'generated-{CUSTOM_VOICE_NAME}.wav',generated_audio.squeeze(0).cpu(),24000)
IPython.display.Audio(f'generated-{CUSTOM_VOICE_NAME}.wav')

```
